Data Preprocessing

```python
# Column - TownToFastChgDriveTime
'''
Here we will replace no with 100 because the values in this column indicated time in minutes which has importance 
so by converting no to 100 will help us to restore the importance.

The Dtype of this column is in object we will convert to int64
'''

train_df['TownToFastChgDriveTime'] = train_df['TownToFastChgDriveTime'].replace('no','100')
train_df['TownToFastChgDriveTime'] = train_df['TownToFastChgDriveTime'].astype(int)

test_df['TownToFastChgDriveTime'] = test_df['TownToFastChgDriveTime'].replace('no','100')
test_df['TownToFastChgDriveTime'] = test_df['TownToFastChgDriveTime'].astype(int)
```

```python
# Column - HwyFastChgDistance
train_df['HwyFastChgDistance'] = train_df['HwyFastChgDistance'].replace('no','200')
train_df['HwyFastChgDistance'] = train_df['HwyFastChgDistance'].astype(int)

test_df['HwyFastChgDistance'] = test_df['HwyFastChgDistance'].replace('no','200')
test_df['HwyFastChgDistance'] = test_df['HwyFastChgDistance'].astype(int)
```

```python
# Column - Race
'''
In race column there are few categories which are very less so we will merge into one of the categorical value to avoid complexity ad doesn't lead
to over fitting.

we will convert Black or African American to Black, 
and Another race, Native Hawaiian or other Pacific Islander, American Indian or Alaska Native, Prefer not to answer merging these categories to Asian.
'''
train_df['race'] = train_df['race'].astype(str)
mer = ['Another race', 'Native Hawaiian or other Pacific Islander', 'American Indian or Alaska Native', 'Prefer not to answer']
train_df['race'] = train_df['race'].replace('Black or African American','Black')
train_df['race'] = train_df['race'].replace(mer,'Asian')

test_df['race'] = test_df['race'].astype(str)
test_df['race'] = test_df['race'].replace('Black or African American','Black')
test_df['race'] = test_df['race'].replace(mer,'Asian')
```

```python
train_df['license'] = train_df['license'].astype(str)
test_df['license'] = test_df['license'].astype(str)
```

```python
# Apply one-hot encoding to the 'race' column
train_df = pd.get_dummies(train_df, columns=['race'], prefix='race',dtype=int)
test_df = pd.get_dummies(test_df, columns=['race'], prefix='race',dtype=int)
```

```python
# Initialize a dictionary to store the number of outliers for each feature
outliers_count = {}

# Calculate and display the number of outliers based on IQR
for feature in num_features:
    Q1 = train_df[feature].quantile(0.25)
    Q3 = train_df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = train_df[(train_df[feature] < lower_bound) | (train_df[feature] > upper_bound)]
    num_outliers = outliers.shape[0]
    outliers_count[feature] = num_outliers
    
    print(f"Number of outliers in {feature}: {num_outliers}")
```

```python
from scipy.stats import skew
# Calculate and display skewness
skewness_values = {}
skewed_features = []

# Calculate and display skewness
for feature in num_features:
    skew_value = skew(train_df[feature])
    skewness_values[feature] = skew_value
    print(f"Skewness of {feature}: {skew_value}")
    
    # Check if the feature is highly skewed
    if skew_value >= 0.025:
        skewed_features.append(feature)

# Display skewed features
print(f"Highly skewed features: {skewed_features}")
```

```python
for feature in skewed_features:
    train_df[f'log_{feature}'] = np.log1p(train_df[feature])
    test_df[f'log_{feature}'] = np.log1p(test_df[feature])
# Drop these features from the dataframe
train_df = train_df.drop(skewed_features, axis=1)
test_df = test_df.drop(skewed_features, axis=1)
```

Feature Selection

```python

from scipy.stats import chi2_contingency
# test for independence

# Create an empty DataFrame to store the results
results = []

for feature in train_df.columns:
    contingency_table = pd.crosstab(train_df['BuyEV'], train_df[feature])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    is_independent = p > 0.025
    results.append({'Feature': feature, 'P-Value': p, 'Independent': is_independent})

# Create a DataFrame from the list of results
results_df = pd.DataFrame(results)

# Set the maximum number of rows to display
pd.set_option('display.max_rows', 100)

# Now, when you print the DataFrame, it will display all rows
print(results_df)
```

```python
# Get a list of independent features
independent_features = results_df.loc[results_df['Independent'] == True, 'Feature'].tolist()

# Drop these features from the dataframe
train_df = train_df.drop(columns=independent_features)
test_df = test_df.drop(columns=independent_features)
```

```python
y = train_df['BuyEV']
X = train_df.drop('BuyEV', axis=1)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
# Initialize the estimator
estimator = RandomForestClassifier(random_state=314)

num_features_to_select = 30
rfe = RFE(estimator, n_features_to_select=num_features_to_select)

# Fit RFE
rfe.fit(X,y)
RFE(estimator=RandomForestClassifier(random_state=314), n_features_to_select=30)
# Get Selected Features
selected_features = X.columns[rfe.support_]

# Get Feature Importances
feature_importances = rfe.estimator_.feature_importances_

# Creating a series with feature names and their scores
importance_series = pd.Series(feature_importances, index=selected_features)

# Sorting the series in descending order to get most important features at the top
sorted_importances = importance_series.sort_values(ascending=False)

# Displaying the sorted importances
print(sorted_importances)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
scaler = MinMaxScaler()


X_train_scaled = scaler.fit_transform(X_train)


X_val_scaled = scaler.transform(X_val)
test_df_scaled = scaler.transform(test_df)
```

Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
model1 = LogisticRegression()

# Train the model
model1.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = model1.predict(X_val_scaled)

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_val, y_pred))
```

Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
# Create a KNN classifier with k=3
k = 2
model2 = KNeighborsClassifier(n_neighbors=k)

# Train the model
model2.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = model2.predict(X_val_scaled)

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_val, y_pred))
```

Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest classifier with 100 trees
model3 = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model3.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = model3.predict(X_val_scaled)

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_val, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
model4 = RandomForestClassifier(random_state=42)
# Define the hyperparameters grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=model4, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Score:", best_score)

# Make predictions on the testing set using the best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val_scaled)

# Evaluate the best model
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_val, y_pred))
```
